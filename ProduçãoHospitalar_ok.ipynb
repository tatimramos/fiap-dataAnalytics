criando função para gerar o gráfico da forma que queremos e setando como parametro da função os dados que irão para o gráfico

```python
def plota_gastos_por_mes(dados):
  axis = dados.T.plot(figsize=(10,6))
  axis.set_ylabel("Gasto Aprovado em Milhões")
  axis.set_xlabel("Ano/Mês")
```

chamando a função e informando como parametro (dados) os dados ordenados por total com os meses filtrados mostrando apenas 3 estados no head

```python
plota_gastos_por_mes(ordenados_por_total[meses_filtrados].head(3))
```

setando limite do eixo y no gráfico

```python
plota_gastos_por_mes(ordenados_por_total[meses_filtrados].head(3))

# como não temos acesso ao axis pela função iremos usar o pyplot
plt.ylim(0,600)
```

```python
# filtrando por menos meses
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[45:70]].head(3))

plt.ylim(0,440)
```

```python
# ultimos 12 meses
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-12:]].head(3))

plt.ylim(0,640)
```

note a diferença entre o gráfico acima e abaixo, no gráfico de baixo parece que SP gasta muitas vezes mais que o Paraná em relação ao gráfico de cima, isso é mentir com dados.\
A melhor prática é deixar o eixo y em 0

```python
# ultimos 12 meses
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-12:]].head(3))

plt.ylim(110,640)
```
